```python
# Dropping the Name and Cabin columns, it is not relevant to the Survival outcome
df.drop(["Name", "Ticket"], axis = 1, inplace = True)
```

```python
# Replacing the values in Embarked points with their full names
df['Embarked'] = df['Embarked'].replace({
    "S" : "Southampton",
    "C" : "Cherbourg",
    "Q" : "Queenstown"
})
df['Embarked'].unique()
```

# Encoding the Categorical Columns

```python
# Using dummy encoding for the Sex column
gender = pd.get_dummies(df['Sex'], dtype = int)
gender
```

```python
df = pd.concat([df, gender], axis = 1)
```

```python
df.drop('Sex', axis = 1, inplace = True)
```

```python
df.head()
```

```python
location = pd.get_dummies(df['Embarked'], dtype = int)
location
```

```python
df = pd.concat([df, location], axis = 1)
```

```python
df.drop("Embarked", axis = 1, inplace = True)
```

```python
df.head()
```

# Separating the features and the target

```python
# Assigning the Features and the Target value
x = df.iloc[:, 1:]
y = df['Survived']
```

```python
x.head(3)
```

```python
# Splitting the dataset 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, shuffle = True, random_state = 42)
```

```python
print(f"The shape of our x_train is {x_train.shape}")
print(f"The shape of our x_test is {x_test.shape}")
print(f"The shape of our y_train is {y_train.shape}")
print(f"The shape of our y_test is {y_test.shape}")
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Model Training

## Logistics Regression model (A binary classification: Survived or not)

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
# Training the logistics Regression with training data
model.fit(x_train, y_train)
print(f"The accuracy score on our training set is: {model.score(x_train, y_train).round(2) * 100}")
```

```python
model.predict(x_test)
```

```python
np.array([y_test])
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,classification_report, ConfusionMatrixDisplay
```

```python
accuracy_score(y_test, model.predict(x_test)).round(2)*100
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test, model.predict(x_test))).plot()
```

# Using Naive Bayes 

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.predict(x_test)
accuracy_score(y_test, nb.predict(x_test))
```
